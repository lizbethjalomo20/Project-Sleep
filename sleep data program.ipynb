## What occupation has the highest blood pressure?

In this problem we're making a list of what occupation has the highest blood pressure starting from 0 through 10. According to the data the job with the worse blood pressure is the nurse with a 116.12 average and the job with the best is an accountant with a 97.32 average. (Kind of ironic that the nurse would be the one with the highest average)

```python
# Split the Blood Pressure column into systolic and diastolic values
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)

# Calculate the average blood pressure for each occupation
avg_blood_pressure_by_occupation = df.groupby('Occupation')[['Systolic', 'Diastolic']].mean()
avg_blood_pressure_by_occupation['Average BP'] = (avg_blood_pressure_by_occupation['Systolic'] + avg_blood_pressure_by_occupation['Diastolic']) / 2

# Sort occupations by the worst average blood pressure
worst_blood_pressure_occupations = avg_blood_pressure_by_occupation.sort_values(by='Average BP', ascending=False)

# Create a ranked list of occupations
ranked_occupations = worst_blood_pressure_occupations.reset_index()

print(ranked_occupations[['Occupation', 'Average BP']])
```
